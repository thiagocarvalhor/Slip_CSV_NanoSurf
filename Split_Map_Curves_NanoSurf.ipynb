```python
import pandas as pd

# Ler o arquivo CSV usando ponto e vírgula como delimitador
df = pd.read_csv("File.csv", header=None, delimiter=';')

# Extrair apenas as colunas 1 e 3 (considerando a indexação iniciando em 0)
dados_iv = df.iloc[:, [0, 2]]

# Renomear as colunas para 'V' e 'I'
dados_iv.columns = ['x', 'y']

# Dividir o DataFrame em 400 partes, cada uma contendo 512 linhas
curvas = [dados_iv[i:i+512]['y'].reset_index(drop=True) for i in range(0, len(dados_iv), 512)]

# Criar um único DataFrame para todas as curvas
df_final = pd.concat(curvas, axis=1)

# Renomear as colunas de corrente para 'y_1', 'y_2', ..., 'y_400'
df_final.columns = [f"y_{i}" for i in range(1, len(curvas) + 1)]

# Adicionar a coluna de voltagem ao DataFrame final
df_final['x'] = dados_iv['x'].iloc[:512].reset_index(drop=True)

# Reordenar as colunas para que 'x' seja a primeira coluna
df_final = df_final[['x'] + [f"y_{i}" for i in range(1, len(curvas) + 1)]]

# Salvar todas as curvas em um único arquivo CSV
df_final.to_csv("curvas_combined_F.csv", index=False)
```
